**<font color='blue'>Read the input water quality files (change the path if needed)<br></font>
<span style="background-color: #FFFF00"><span style="color:#f00">Note that</span></span> all results are generated with detrended forcings.**

```python
basin_name = ['Mercer', 'Thornton', 'Issaquah']
basin_nameL = ['mercer', 'thornton', 'issaquah']
outputfilepath = 'D:\\Box Sync\\WQ-PAPER\\Figures\\'  

N = 3
colors = ['b', 'black', 'r', 'c', 'm', 'y', 'k'] 
#colors = np.random.rand(N)

linestyles = ['-', '--', '-', '-', '-', '-']

scenarios = ['1883','2002','2050', 'TairRcp45', 'TairRcp85', 'PptRcp85', 'Combined']
strings = ['LC1883','Baseline' ,'LC2050', 'RCP4.5(T))', 'RCP8.5(T))', 'Combined']


# the period of interest for daily pollutant load plots
startdate = datetime.datetime(1993,1,1)
enddate = datetime.datetime(2006,12,31)

mainpath = {}
# define the 2-dimensional dictionary with multiple keys per value
simflowfilepath = collections.defaultdict(dict)
precipfilepath = collections.defaultdict(dict)

# read in 3-hourly data file path
for i, n in enumerate(basin_name):
    mainpath[n] = 'D:\\Dropbox\\Python_Scripts\\'+str(basin_name[i])
    for j, s in enumerate(scenarios):
        if (s=='1883' or s=='2002' or s=='2050'):
            simflowfilepath[n][s] = str(mainpath[n])+'\\LULC_Scenarios\\'+str(s)+'\\output\\Streamflow.Only'
            precipfilepath[n][s] = str(mainpath[n])+'\\LULC_Scenarios\\'+str(s)+'\\output\\Mass.Balance'
        else:
            simflowfilepath[n][s] = str(mainpath[n])+'\\Climate_Scenarios\\'+str(s)+'\\output\\Streamflow.Only'
            precipfilepath[n][s] = str(mainpath[n])+'\\Climate_Scenarios\\'+str(s)+'\\output\\Mass.Balance'
```

<font color='blue'>**Read the input water quality files (change the path if needed)**</font><br>

```python
tempsimdates = collections.defaultdict(list)
simflowdata = collections.defaultdict(dict)
precipdata = collections.defaultdict(dict)
simETdata = collections.defaultdict(dict)

# dates for all other files but Temperature
tempsimdates = np.genfromtxt(simflowfilepath['Mercer']['1883'], dtype=str, skiprows=2, usecols=0)
simdates = []
for i in range(len(tempsimdates)):
    simdates.append(datetime.datetime.strptime(tempsimdates[i],'%m/%d/%Y-%H:%M:%S'))
    
    
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        simflowdata[n][s] = np.genfromtxt(simflowfilepath[n][s], skiprows=2, usecols=[1])  
        simETdata[n][s] = np.genfromtxt(precipfilepath[n][s], dtype=float, skiprows=24, usecols=[8])
        precipdata[n][s] = np.genfromtxt(precipfilepath[n][s], dtype=float, skiprows=24, usecols=[9])
```

<font color='blue'>**Calculate the annual sum of load.**</font>

```python
DailyFlowRate = collections.defaultdict(dict)
DailyET = collections.defaultdict(dict)
DailyPrecip = collections.defaultdict(dict)
MonthlyFlowRate = collections.defaultdict(dict)
AnnualFlowRate = collections.defaultdict(dict)
MonthlyET = collections.defaultdict(dict)
MonthlyPrecip = collections.defaultdict(dict)
monthFlow = collections.defaultdict(dict)

for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        DailyFlowRate[n][s] = pd.Series(simflowdata[n][s], index=simdates).resample('D', how='sum')/3/3600
        DailyET[n][s] = pd.Series(simETdata[n][s], index=simdates).resample('D', how='sum') # in meters
        DailyPrecip[n][s] = pd.Series(precipdata[n][s], index=simdates).resample('D', how='sum') # in meters

# Truncate the data into desired time frame
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        DailyFlowRate[n][s] = DailyFlowRate[n][s].truncate(before=startdate, after=enddate)
        DailyET[n][s] = DailyET[n][s].truncate(before=startdate, after=enddate)
        DailyPrecip[n][s] = DailyPrecip[n][s].truncate(before=startdate, after=enddate)

# average monthly streamflow for each year
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        MonthlyFlowRate[n][s] = DailyFlowRate[n][s].resample('M', how='mean')
        
        
# average monthly streamflow over a period
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        monthFlow[n][s] = DailyFlowRate[n][s].groupby(lambda x: x.month).mean()
        MonthlyET[n][s] = DailyET[n][s].groupby(lambda x: x.month).mean()
        MonthlyPrecip[n][s] = DailyPrecip[n][s].groupby(lambda x: x.month).mean()

# Annual Sum Streamflow
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        AnnualFlowRate[n][s] = DailyFlowRate[n][s].resample('A', how='mean')
```

<font color='blue'>Plot **Annual avg. **streamflow</font>

```python
fig=plt.figure(figsize=(6, 9),dpi=400)

for i, n in enumerate(basin_name):
    ax = plt.subplot(3, 1, i+1)
    fig.subplots_adjust(hspace = 0.3, wspace = 0.1)
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    # add the catchment name to the figure
    ax.text(.95,.02, str(n), fontsize=12, horizontalalignment='right', transform=ax.transAxes)
    for (j, s), c, t, l in zip(enumerate(scenarios), colors, strings, linestyles):
        AnnualFlowRate[n][s].plot(color=c, label=str(t), linewidth=2.0, linestyle=l)
    plt.ylabel('streamflow ($m^{3}/s$)', fontsize=12.)
    if (i == 2):
        plt.xlabel('year', fontsize=12.)
    #grid(True)
    if (i == 0):
        ax.legend(loc='best', bbox_to_anchor=(1.05, 0.99))

#outfig1 = str(outputfilepath) + 'Scenarios-Annual-Streamflow.png'      
#savefig(str(outfig1), dpi=500, format = 'png', transparent='True')
```

<font color='blue'>Plot **average monthly flow** over the period of interest</font>

```python
# Reorganize the data set to conform to the wateryear (ie. Oct-Sep)
for i, n in enumerate(basin_name):
    for j, s in enumerate(scenarios):
        monthFlow[n][s] = monthFlow[n][s][9:].append(monthFlow[n][s][:9])
```
