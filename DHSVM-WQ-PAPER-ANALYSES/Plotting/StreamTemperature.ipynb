```python
basin_name = ['Thornton', 'Mercer', 'Issaquah']
mainpath = {}
simtemperaturepath = {}
obstemperaturepath = {}

for i, n in enumerate(basin_name):
    mainpath[n] = 'D:\\Dropbox\\Python_Scripts\\'+str(basin_name[i])+'\\water_temp\\'
    simtemperaturepath[n] = str(mainpath[n])+'\\sim2.txt'
    obstemperaturepath[n] = str(mainpath[n])+'\\obs2.txt'
```

**<font color='blue'>Read the data in. </font>**

```python
simtempdata = {}
obstempdata = {}
tempsimdate = {}
tempobsdate = {}

# Read flow sim and obs dates &  data
for i, n in enumerate(basin_name):
    simtempdata[n] = np.genfromtxt(simtemperaturepath[n], usecols=[1])
    obstempdata[n] = np.genfromtxt(obstemperaturepath[n], usecols=[1])
    tempsimdate[n] = np.genfromtxt(simtemperaturepath[n], dtype=str, skiprows=0, usecols=0)
    tempobsdate[n] = np.genfromtxt(obstemperaturepath[n], dtype=str, skiprows=0, usecols =0)
```

<font color='blue'>**Sort and match the obs. and sim. flow files based on dates.**<br></font>

```python
# Convert the string format dates into a datetime object
simdate = collections.defaultdict(list)

for i, n in enumerate(basin_name):
    for j in range(len(tempsimdate[n])):
        simdate[n].append(datetime.datetime.strptime(tempsimdate[n][j],'%m/%d/%Y-%H:%M'))
# sort the dates in an acending order
for i, n in enumerate(basin_name):
    simdate[n].sort()

# Convert the string format dates into a datetime object
obsdate = collections.defaultdict(list)
for i, n in enumerate(basin_name):
    for j in range(len(tempobsdate[n])):
        obsdate[n].append(datetime.datetime.strptime(tempobsdate[n][j],'%m/%d/%Y-%H:%M'))
for i, n in enumerate(basin_name):
    obsdate[n].sort()
```

```python
Fsimdate = collections.defaultdict(list)
Fobsdate = collections.defaultdict(list)
FsimTemp = collections.defaultdict(list)
FobsTemp = collections.defaultdict(list)

for j, n in enumerate(basin_name):
    for i in range(len(obsdate[n])):
        obs = dates.date2num(obsdate[n][i])
        junk = date2num(simdate[n])
        idx = (np.abs(junk-obs)).argmin()
        if (idx-1 > 0 and (idx+1)< len(simdate[n])):
            Fsimdate[n].append(simdate[n][idx])
            Fobsdate[n].append(obsdate[n][i])
            FsimTemp[n].append(simtempdata[n][idx])
            FobsTemp[n].append(obstempdata[n][i])
```

```python
for i, n in enumerate(basin_name):
    ssim[n] = pd.Series(FsimTemp[n], index=Fsimdate[n])
    sobs[n] = pd.Series(FobsTemp[n], index=Fobsdate[n])

startdate = datetime.datetime(1993,1,1)
enddate = datetime.datetime(2008,12,31)
for j, n in enumerate(basin_name):
    ssim[n] = ssim[n].truncate(before=startdate, after=enddate)
    sobs[n] = sobs[n].truncate(before=startdate, after=enddate)
    FsimTemp[n] = ssim[n].tolist()
    FobsTemp[n] = sobs[n].tolist()
```

<font color='blue'>**Basic Stats: min and max and avg.** </font>

```python
HourlyTemp = collections.defaultdict(list)
Temp = collections.defaultdict(list)

for i, n in enumerate(basin_name):
    HourlyTemp[n] = pd.Series(simtempdata[n], index=simdate[n])

print "min, max, avg:"
for i, n in enumerate(basin_name):
    Temp[n] = HourlyTemp[n].truncate(before=startdate, after=enddate)

for i, n in enumerate(basin_name):
    print str(n), "%.2f %.2f %.2f" % (min(Temp[n]), max(Temp[n]), average(Temp[n]))
    
```
