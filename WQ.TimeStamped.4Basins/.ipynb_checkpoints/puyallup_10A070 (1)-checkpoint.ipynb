```python
import matplotlib.pyplot as plt
from matplotlib.finance import quotes_historical_yahoo
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
import numpy as np
import pandas as pd
import datetime as dt
from matplotlib.dates import date2num, num2date 
from matplotlib.pyplot import figure, show
from matplotlib import scale as mscale
from matplotlib import transforms as mtransforms
import pylab
import matplotlib
import os
%matplotlib inline
#import scikits.timeseries as ts
#import scikits.timeseries.lib.plotlib as tpl

os.chdir('D:\\station_0631')

data = np.loadtxt('obs.txt')
data1 = np.genfromtxt('obs.txt',dtype='int')
year = data1[:,0]
month = data1[:,1]
day = data1[:,2]
hour = data1[:,3]
minute = data1[:,4]
obs = data[:,5]
time=np.zeros([data.shape[0],1])
#print len(time)
#print len(sim)
for i in range(0,data.shape[0]):
    a= year[i]
    b= month[i]
    c= day[i]
    d= hour[i]
    e= minute[i]
    f = dt.datetime(a,b,c,d,e)
    time[i] = date2num(f)
    i+=1
numobs=np.arange(data.shape[0])

data2 = np.loadtxt('sim.txt')
data3 = np.genfromtxt('sim.txt',dtype='int')
year1 = data3[:,0]
month1 = data3[:,1]
day1 = data3[:,2]
hour1 = data3[:,3]
minute1 = data3[:,4]
sim = data2[:,5]
time1=np.zeros([data2.shape[0],1])
#print len(time)
#print len(sim)
for i in range(0,data2.shape[0]):
    a1= year1[i]
    b1= month1[i]
    c1= day1[i]
    d1= hour1[i]
    e1= minute1[i]
    f1 = dt.datetime(a1,b1,c1,d1,e1)
    time1[i] = date2num(f1)
    i+=1
numsim=np.arange(data2.shape[0])
#obs1=np.zeros([data.shape[0],2])
#obs1=[time,obs]
##obs1[:,0]= time
##obs1[:,1]= obs
#print obs1

def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
array= time1
#value= time[3]
#print value
#print(find_nearest(array, value))

sim_obs=np.zeros([data.shape[0],4])
for i in range(0,data.shape[0]):
    value= time[i]
    sim_obs[i,0]=time[i]
    sim_obs[i,3]=obs[i]
    idx = (np.abs(array-value)).argmin()
    sim_obs[i,1]=array[idx]
    sim_obs[i,2]=sim[idx]
    i+=1
#print sim_obs[:,0]
#fig,(ax) = plt.plot()
bias=np.sum(sim_obs[:,2]-sim_obs[:,3])/data.shape[0]
#SSres=sum((sim_obs[:,2]-sim_obs[:,3])**2)
#SStot=sum((sim_obs[:,3]-np.mean(sim_obs[:,3]))**2)
#R2=1-SSres/SStot
tempr1=0
tempr2=0
tempr3=0
tempr4=0
for i in range(0,sim_obs[:,0].shape[0]):
    tempr1+=(sim_obs[i,2]-np.mean(sim_obs[:,2]))*(sim_obs[i,3]-np.mean(sim_obs[:,3]))
    tempr2+=(sim_obs[i,2]-np.mean(sim_obs[:,2]))**2
    tempr3+=(sim_obs[i,3]-np.mean(sim_obs[:,3]))**2
    tempr4+=(sim_obs[i,3]-sim_obs[i,2])**2
    i+=1
R2=1-tempr4/tempr3
RMSE=(sum((sim_obs[:,2]-sim_obs[:,3])**2)/data.shape[0])**0.5
#print sim_obs[:,2].reshape(data.shape[0],1)
#print sim_obs[:,0]
#print sim_obs[:,3]
#np.savetxt('sim_test.txt',sim_obs[:,2])
#np.savetxt('obs_test.txt',sim_obs[:,3])
print bias,R2,RMSE
bias_1="{:.2f}".format(bias)
R2_1="{:.2f}".format(R2)
RMSE_1="{:.2f}".format(RMSE)
#print bias1
#fig,(ax) = plt.plot()
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.plot_date(sim_obs[:,0], sim_obs[:,2], 'r',label='Sim')
ax.plot_date(sim_obs[:,0], sim_obs[:,3], 'b',label='Obs')
plt.gcf().autofmt_xdate()
legend = ax.legend(loc='upper right', shadow=True)
plt.title('Water Temperature of Puyallup (10A070)',fontsize=16)
pylab.ylim([1,21])
ax.set_xlabel('time',fontsize=14, x=0.5, y=0)
ax.set_ylabel('Water Temperature (degC)',fontsize=14)
matplotlib.rc('xtick', labelsize=14)
matplotlib.rc('ytick', labelsize=14)
#ax.text(0.1,0.3,'R2')
plt.text(729750,19,'$R^2= $'+str(R2_1),fontsize=14)
plt.text(729750,18,'$RMSE= $'+str(RMSE_1),fontsize=14)
plt.text(729750,17,'$Bias= $'+str(bias_1),fontsize=14)
#figure(num=None, figsize=(20, 12), dpi=200, facecolor='w', edgecolor='k')
#plt.show()
plt.savefig('sim_obs.png',dpi=500,format = 'png',transparent='True')
```
