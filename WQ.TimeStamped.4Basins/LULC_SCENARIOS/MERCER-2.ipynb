```python
# Enter file path for sim. and obs. flow files
basin_name = "Mercer"
basin_nameL = "mercer"
mainpath = 'D:\\Dropbox\\Python_Scripts\\'+str(basin_name)
print "mainpath is:", mainpath
print ""

years = ['1883','2002','2027','2050']
markers = ["s","*", "D", "v", "h", "p", "H", "o"]
colors = ["Black", "Blue", "Green", "Red", "Lime", "Magenta", "Aqua", "DarkOrange"] 
linestyles = ['-', '-', '-', '--', '-', '-', '-', '--']

# the period of interest for daily pollutant load plots
startdate = datetime.datetime(1993,1,1)
enddate = datetime.datetime(2008,12,31)

simflowpath = {}
simflowdata = {}
dailyprecippath = {}
simETdata = {}
simETpath = {}

for i, year in enumerate(years):
    simflowpath[year]= str(mainpath) +'\\LULC_Scenarios\\' + str(year) + '\\output\\daily_flow.txt'
    simETpath[year] = str(mainpath) + '\\LULC_Scenarios\\' + str(year) + '\\output\\daily_ET.txt'
    print simflowpath[year]
obsflowpath = str(mainpath) + '\\streamflow.12120000.cms.txt'
dailyprecippath = str(mainpath) + '\\LULC_Scenarios\\1883\\output\\daily_precip.txt'

# Read flow sim and obs dates &  data
for i, year in enumerate(years):
    simflowdata[year] = np.genfromtxt(simflowpath[year], usecols=[1])  
    simETdata[year] = np.genfromtxt(simETpath[year], usecols=[1])
obsflowdata = np.genfromtxt(obsflowpath, usecols=[1])

tempsimflowdate = np.genfromtxt(simflowpath['1883'], dtype = str, skiprows = 0, usecols = 0)
tempobsflowdate = np.genfromtxt(obsflowpath, dtype = str, skiprows = 0, usecols = 0)

dailyprecipdata = np.genfromtxt(dailyprecippath, usecols=[1])
```

<font color='blue'>Read the input **water quality** files <br>
</font>

```python
simsedfilepath = {}
simotherfilepath = {}
simOrgPfilepath = {}
simSolPfilepath = {}
sim3hrflowfilepath = {}

simseddata = {}
simotherdata = {}
simOrgnPdata = {}
simSolPdata = {}
simTPdata = {}
sim3hrflow = {}

for i, year in enumerate(years):
    sim3hrflowfilepath[year] = str(mainpath) +'\\LULC_Scenarios\\'+str(year)+'\\output\\Streamflow.Only'
    simsedfilepath[year] = str(mainpath) +'\\LULC_Scenarios\\'+str(year)+'\\output\\Sed.Flow.Only'
    simotherfilepath[year] = str(mainpath) +'\\LULC_Scenarios\\'+str(year)+'\\output\\Other.Conc.Only'

# Enter file path for observed precipitation files
precipfilepath = str(mainpath) +'\\LULC_Scenarios\\1883\\output\\Mass.Balance'

# Read dates of sims (continuous)
tempsimdates = np.genfromtxt(simsedfilepath['1883'], dtype=str, skiprows=2, usecols=0)
simdates = []
for i in range(len(tempsimdates)):
    simdates.append(datetime.datetime.strptime(tempsimdates[i],'%m/%d/%Y-%H:%M:%S'))

# Read 3-hourly flow sims (use the 2nd col)
for i, year in enumerate(years):
    sim3hrflow[year] = np.genfromtxt(sim3hrflowfilepath[year], skiprows=2, usecols=[1])  
    
# Read sediment sims (use the 2nd col)
for i, year in enumerate(years):
    simseddata[year] = np.genfromtxt(simsedfilepath[year], skiprows=2, usecols=[2])  
    simseddata[year] = simseddata[year] + 2

# Read other data
for i, year in enumerate(years):
    simotherdata[year] = np.genfromtxt(simotherfilepath[year], skiprows=2, usecols=[1])
    simotherdata[year] = simotherdata[year] + 20

# Read obs. precipitation data
precipdata = np.genfromtxt(precipfilepath, dtype=float, skiprows=24, usecols=[9])

# Read TP data
for i, year in enumerate(years):
    simOrgPfilepath[year] = str(mainpath) + '\\LULC_Scenarios\\' + str(year) + '\\output\\OrgP.Conc.Only'
    simSolPfilepath[year] = str(mainpath) + '\\LULC_Scenarios\\' + str(year) + '\\output\\solP.Conc.Only'

for i, year in enumerate(years):
    simOrgnPdata[year] = np.genfromtxt(simOrgPfilepath[year], skiprows=2, usecols=[1])
    simSolPdata[year] = np.genfromtxt(simSolPfilepath[year], skiprows=2, usecols=[1])
    simTPdata[year] = simSolPdata[year] + simOrgnPdata[year]
```

```python
DailySedLoad = {}
MonthlySedLoad = {}
DailyColiformLoad = {}
MonthlyColiformLoad = {}
DailyTPLoad = {}
MonthlyTPLoad = {}
MonthlyFlow = {}
junk = {}
for i, year in enumerate(years):
    DailySedLoad[year] = pd.Series(simseddata[year]*sim3hrflow[year]/(10**6), index=simdates).resample('D', how='sum')
    DailyColiformLoad[year] = pd.Series(simotherdata[year]*sim3hrflow[year]/(10**5), index=simdates).resample('D', how='sum')
    DailyTPLoad[year] = pd.Series(simTPdata[year]*sim3hrflow[year]/(10**3), index=simdates).resample('D', how='sum')

# Truncate the data into desired time frame
for i, year in enumerate(years):
    DailySedLoad[year] = DailySedLoad[year].truncate(before=startdate, after=enddate)
    DailyColiformLoad[year] = DailyColiformLoad[year].truncate(before=startdate, after=enddate)
    DailyTPLoad[year] = DailyTPLoad[year].truncate(before=startdate, after=enddate)
    junk[year] = pd.Series(sim3hrflow[year], index=simdates).truncate(before=startdate, after=enddate)

for i, year in enumerate(years):
    MonthlySedLoad[year] = DailySedLoad[year].resample('M', how='sum')
    MonthlyColiformLoad[year] = DailyColiformLoad[year].resample('M', how='sum')
    MonthlyTPLoad[year] = DailyTPLoad[year].resample('M', how='sum')
    MonthlyFlow[year] = junk[year].resample('M', how='sum')
```
