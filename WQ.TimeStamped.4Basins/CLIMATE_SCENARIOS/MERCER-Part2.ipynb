<font color='blue'>Read the input **streamflow** files (change the path if needed)<br>
<span style="background-color: #FFFF00">**Attention**: the last date in obs. file **MUST NOT** exceed the last date of sim.</span></font>

```python
# Enter file path for sim. and obs. flow files
basin_name = "Mercer"
basin_nameL = "mercer"
subdir_name = "Climate_Scenarios"
mainpath = 'D:\\Dropbox\\Python_Scripts\\'+str(basin_name)
print "mainpath is:", mainpath
print ""

years = ['PptRcp85', 'TairRcp85', 'hist']
markers = ["s","*", "D", "v", "h", "p", "H", "o"]
colors = ["Aqua", "Blue", "Green", "Red", "Lime", "Magenta", "Black", "DarkOrange"] 
linestyles = ['-', '-', '-', '-', '-', '-', '-', '-']

# the period of interest for daily pollutant load plots
startdate = datetime.datetime(1993,1,1)
enddate = datetime.datetime(2008,12,31)

simflowpath = {}
simflowdata = {}
dailyprecippath = {}
dailyprecipdata = {}
simETdata = {}
simETpath = {}

for i, year in enumerate(years):
    simflowpath[year]= str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\daily_flow.txt'
    simETpath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\daily_ET.txt'
    dailyprecippath = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\daily_precip.txt'
    print simflowpath[year]
    
obsflowpath = str(mainpath)+'\\streamflow.12120000.cms.txt'

# Read flow sim and obs dates &  data
for i, year in enumerate(years):
    simflowdata[year] = np.genfromtxt(simflowpath[year], usecols=[1])  
    simETdata[year] = np.genfromtxt(simETpath[year], usecols=[1])
obsflowdata = np.genfromtxt(obsflowpath, usecols=[1])

tempsimflowdate = np.genfromtxt(simflowpath['PptRcp85'], dtype = str, skiprows = 0, usecols = 0)
tempobsflowdate = np.genfromtxt(obsflowpath, dtype = str, skiprows = 0, usecols = 0)

for i, year in enumerate(years):
    dailyprecipdata[year] = np.genfromtxt(dailyprecippath, usecols=[1])
```

<font color='blue'>Read the input **water quality** files <br>
</font>

```python
simsedfilepath = {}
simotherfilepath = {}
simOrgPfilepath = {}
simSolPfilepath = {}
sim3hrflowfilepath = {}
precipfilepath = {}
precipdata = {}

simseddata = {}
simotherdata = {}
simOrgnPdata = {}
simSolPdata = {}
simTPdata = {}
sim3hrflow = {}

for i, year in enumerate(years):
    sim3hrflowfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\Streamflow.Only'
    simsedfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\Sed.Flow.Only'
    simotherfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\Other.Conc.Only'
    precipfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\Mass.Balance'

# Read dates of sims (continuous)
tempsimdates = np.genfromtxt(simsedfilepath['hist'], dtype=str, skiprows=2, usecols=0)
simdates = []
for i in range(len(tempsimdates)):
    simdates.append(datetime.datetime.strptime(tempsimdates[i],'%m/%d/%Y-%H:%M:%S'))

# Read 3-hourly flow sims (use the 2nd col)
for i, year in enumerate(years):
    sim3hrflow[year] = np.genfromtxt(sim3hrflowfilepath[year], skiprows=2, usecols=[1])  
    
# Read sediment sims (use the 2nd col)
for i, year in enumerate(years):
    simseddata[year] = np.genfromtxt(simsedfilepath[year], skiprows=2, usecols=[2])  
    simseddata[year] = simseddata[year] + 2

# Read other data
for i, year in enumerate(years):
    simotherdata[year] = np.genfromtxt(simotherfilepath[year], skiprows=2, usecols=[1])
    simotherdata[year] = simotherdata[year] + 20
    precipdata[year] = np.genfromtxt(precipfilepath[year], dtype=float, skiprows=24, usecols=[9])


# Read TP data
for i, year in enumerate(years):
    simOrgPfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\OrgP.Conc.Only'
    simSolPfilepath[year] = str(mainpath)+'\\'+str(subdir_name)+'\\'+str(year)+'\\output\\solP.Conc.Only'

for i, year in enumerate(years):
    simOrgnPdata[year] = np.genfromtxt(simOrgPfilepath[year], skiprows=2, usecols=[1])
    simSolPdata[year] = np.genfromtxt(simSolPfilepath[year], skiprows=2, usecols=[1])
    simTPdata[year] = simSolPdata[year] + simOrgnPdata[year]
```

```python
DailySedLoad = {}
MonthlySedLoad = {}
DailyColiformLoad = {}
MonthlyColiformLoad = {}
DailyTPLoad = {}
MonthlyTPLoad = {}
MonthlyFlow = {}
junk = {}
for i, year in enumerate(years):
    DailySedLoad[year] = pd.Series(simseddata[year]*sim3hrflow[year]/(10**6), index=simdates).resample('D', how='sum')
    DailyColiformLoad[year] = pd.Series(simotherdata[year]*sim3hrflow[year]/(10**5), index=simdates).resample('D', how='sum')
    DailyTPLoad[year] = pd.Series(simTPdata[year]*sim3hrflow[year]/(10**3), index=simdates).resample('D', how='sum')

# Truncate the data into desired time frame
for i, year in enumerate(years):
    DailySedLoad[year] = DailySedLoad[year].truncate(before=startdate, after=enddate)
    DailyColiformLoad[year] = DailyColiformLoad[year].truncate(before=startdate, after=enddate)
    DailyTPLoad[year] = DailyTPLoad[year].truncate(before=startdate, after=enddate)
    junk[year] = pd.Series(sim3hrflow[year], index=simdates).truncate(before=startdate, after=enddate)

for i, year in enumerate(years):
    MonthlySedLoad[year] = DailySedLoad[year].resample('M', how='sum')
    MonthlyColiformLoad[year] = DailyColiformLoad[year].resample('M', how='sum')
    MonthlyTPLoad[year] = DailyTPLoad[year].resample('M', how='sum')
    MonthlyFlow[year] = junk[year].resample('M', how='sum')
```

<font color='blue'>Plot the relationship between **monthly** flow and monthly pollutant load</font>

```python
fig=plt.figure(figsize=(7,5),dpi=400)
ax = plt.subplot()
# log-normal probability plot
nn = len(DailySedLoad['hist'])
dd = np.linspace(1, nn, num=nn)/(nn+1)
yy  = norm.ppf(dd,0,1);

# create the axis ticks
p  = [0.001, 0.01, 0.05, 0.2, 0.5, 0.75, 0.90, 0.98, 0.999];
# relate var "p" with "y"
tick  = norm.ppf(p,0,1);
label = ['0.001','0.01','0.05','0.2','0.5','0.75','0.90','0.98','0.999'];
# recast the p to (0,1)

# sort the data in an ascending order
for i, year in enumerate(years):
    DailySedLoad[year].sort()
    
# plot with all points within 9-hour interval of obs collection time.
for (i, year), c, m in zip(enumerate(years), colors, markers):
    ax.plot(yy, DailySedLoad[year], color=c, marker=m, markersize=4, markeredgewidth=0.0, linestyle ='None', label=str(year))
#ax.plot(y, sobs,'b*', label='obs')

# use numpoints option so the markeres don't appear twice in one legend
ax.legend(loc='best',numpoints = 1)
plt.ylabel('load ($10^{3}kg$)', fontsize=14.)
plt.xlabel('cumulative prob.', fontsize=14.)
plt.xlim([min(yy), max(yy)])

ax = gca()
ax.xaxis.set_major_locator(ticker.FixedLocator(tick))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(label))
grid(True)
# save the figure to the folder
outputfilepath = str(mainpath)+'\\'+str(subdir_name)+'\\Daily_Sediment_load_cdf.png'
savefig(str(outputfilepath),dpi=500,format = 'png',transparent='True')
```
