**<font color='blue'>Read the NSW.Only files from old and new model versions.</font>**

```python
mainpath = []
oldvpath = []
newvpath = []
iswpath = []
old_NSW = []
new_NSW = []
ISW = []
date = []

mainpath = 'D:\\ModelUpdateTests\\test1B\\'
oldvpath = str(mainpath)+'\\NSW-old.Only'
newvpath = str(mainpath)+'\\NSW-new.Only'
iswpath = str(mainpath)+'ISW.Only'

# use the NSW data for Seg **
m = 1
n = 51
```

###<font color='red'>Plot the segment n</font>###

```python
fdate = []
old_NSW = np.genfromtxt(oldvpath, dtype=float, skiprows=2, usecols=[n])
new_NSW = np.genfromtxt(newvpath, dtype=float, skiprows=2, usecols=[n])
ISW = np.genfromtxt(iswpath, dtype=float, skiprows=2, usecols=[n])

date = np.genfromtxt(oldvpath, dtype=str, skiprows=2, usecols=[0])

# Convert the string format dates into a datetime object
for i in range(len(date)):
    fdate.append(datetime.datetime.strptime(date[i], '%m.%d.%Y-%H:%M:%S'))

old = []
new = []
isw = []    
old = pd.Series(old_NSW, index=fdate)
new = pd.Series(new_NSW, index=fdate)
isw = pd.Series(ISW, index=fdate)
```

<font color='blue'>**Scatterplot of NSW @ Seg.n** </font>

```python
plt.scatter(old, new)
a = old.max()
b = old.min()
plt.xlim([a, b])
plt.ylim([a, b])
plt.xlabel('orig NSW')
plt.ylabel('new NSW')
```

<font color='blue'>**Plot the time series of NSW @ Seg.** </font>

```python
fig = plt.figure(figsize=(12, 3), dpi=400)
plt.title('NSW comparison at segment %s' % str(n))
old.plot(label='orig', color='red', linewidth=2.)
new.plot(label='new', color='blue', linewidth=2.)
grid(True)
plt.legend(loc='best')    
outputfilepath = str(mainpath) + 'NSW_COMP.png'    
#savefig(str(outputfilepath),dpi=300,format='png',transparent='True')
```

```python
fig = plt.figure(figsize=(12, 3), dpi=400)
plt.title('ISW comparison at segment %s' % str(n))
isw.plot(label='orig', color='red', linewidth=2.)
grid(True)
plt.legend(loc='best')    
```
