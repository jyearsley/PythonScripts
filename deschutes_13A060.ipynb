```python
import matplotlib.pyplot as plt
from matplotlib.finance import quotes_historical_yahoo
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
import numpy as np
import pandas as pd
import datetime as dt
from matplotlib.dates import date2num, num2date 
from matplotlib.pyplot import figure, show
from matplotlib import scale as mscale
from matplotlib import transforms as mtransforms
import pylab
import matplotlib
%matplotlib inline
#import scikits.timeseries as ts
#import scikits.timeseries.lib.plotlib as tpl

filepath='/raid/caoqian/DHSVM3.1.2_PugetSound_Run/plot/deschutes/13A060/results/1.txt'
#dates = dt.datetime.strptime("07/27/2012","%m/%d/%Y")
#print dates

data = np.loadtxt(filepath)
data1 = np.genfromtxt(filepath,dtype='int')
year = data1[:,0]
month = data1[:,1]
day = data1[:,2]
sim = data[:,3]
obs = data[:,4]
time=np.zeros([data.shape[0],1])
#print len(time)
#print len(sim)
for i in range(0,data.shape[0]):
    a= year[i]
    b= month[i]
    c= day[i]
    d = dt.date(a,b,c)
    time[i] = date2num(d)
    i+=1
time1=np.arange(data.shape[0])
#series = ts.fill_missing_dates(sim)

y1=sim
y2=obs
x=time
fig,(ax,ax2,ax3,ax4,ax5,ax6) = plt.subplots(1, 6, sharey=True,figsize=(14,6))
#ax0 = fig.add_subplot(111)

# plot the same data on both axes

ax.plot_date(x, y1, 'r')
ax2.plot_date(x, y1, 'r')
ax3.plot_date(x, y1, 'r')
ax4.plot_date(x, y1, 'r')
ax5.plot_date(x, y1, 'r')
ax6.plot_date(x, y1, 'r',label='Sim')
ax.plot_date(x, y2, 'b')
ax2.plot_date(x, y2, 'b')
ax3.plot_date(x, y2, 'b')
ax4.plot_date(x, y2, 'b')
ax5.plot_date(x, y2, 'b')
ax6.plot_date(x, y2, 'b',label='Obs')

# zoom-in / limit the view to different portions of the data
filepath1='/raid/caoqian/DHSVM3.1.2_PugetSound_Run/plot/deschutes/13A060/date_tmp13A060.txt'
datase = np.genfromtxt(filepath1,dtype='int')
yearse = datase[:,0]
monthse = datase[:,1]
dayse = datase[:,2]
timese=np.zeros([datase.shape[0],1])
for i in range(0,datase.shape[0]):
    a= yearse[i]
    b= monthse[i]
    c= dayse[i]
    d = dt.date(a,b,c)
    timese[i] = date2num(d)
    i+=1
ax.set_xlim(timese[0],timese[1]) 
ax2.set_xlim(timese[2],timese[3]) 
ax3.set_xlim(timese[4],timese[5])
ax4.set_xlim(timese[6],timese[7]) 
ax5.set_xlim(timese[8],timese[9]) 
ax6.set_xlim(timese[10],timese[11])
# hide the spines between ax and ax2

ax.spines['right'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax4.spines['left'].set_visible(False)
ax4.spines['right'].set_visible(False)
ax5.spines['left'].set_visible(False)
ax5.spines['right'].set_visible(False)
ax6.spines['left'].set_visible(False)

ax.yaxis.tick_left()
ax.tick_params(labeltop='off') # don't put tick labels at the top
#for ax in [ax2, ax3, ax4, ax5]:
    #plt.step(ax2.get_yticklabels()[-1], visible=False)
    #ax.set_yticks([])
ax6.yaxis.tick_right()
for m in [ax2, ax3, ax4, ax5]:
    m.tick_params(axis='y', left='off',right='off')
    
# Make the spacing between the two axes a bit smaller
plt.subplots_adjust(wspace=0.05)

d = .015 # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((1-d,1+d),(-d,+d), **kwargs) # top-left diagonal
ax.plot((1-d,1+d),(1-d,1+d), **kwargs) # bottom-left diagonal

kwargs.update(transform=ax2.transAxes) # switch to the bottom axes
ax2.plot((-d,d),(-d,+d), **kwargs) # top-right diagonal
ax2.plot((-d,d),(1-d,1+d), **kwargs) # bottom-right diagonal
ax2.plot((1-d,1+d),(-d,+d), **kwargs) # top-left diagonal
ax2.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax3.transAxes) # switch to the bottom axes
ax3.plot((-d,d),(-d,+d), **kwargs) # top-right diagonal
ax3.plot((-d,d),(1-d,1+d), **kwargs) # bottom-right diagonal
ax3.plot((1-d,1+d),(-d,+d), **kwargs) # top-left diagonal
ax3.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax4.transAxes) # switch to the bottom axes
ax4.plot((-d,d),(-d,+d), **kwargs) # top-right diagonal
ax4.plot((-d,d),(1-d,1+d), **kwargs) # bottom-right diagonal
ax4.plot((1-d,1+d),(-d,+d), **kwargs) # top-left diagonal
ax4.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax5.transAxes) # switch to the bottom axes
ax5.plot((-d,d),(-d,+d), **kwargs) # top-right diagonal
ax5.plot((-d,d),(1-d,1+d), **kwargs) # bottom-right diagonal
ax5.plot((1-d,1+d),(-d,+d), **kwargs) # top-left diagonal
ax5.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax6.transAxes) # switch to the bottom axes
ax6.plot((-d,d),(-d,+d), **kwargs) # top-right diagonal
ax6.plot((-d,d),(1-d,1+d), **kwargs) # bottom-right diagonal

plt.gcf().autofmt_xdate()
#locator = AutoDateLocator()
#locator.intervald[MONTHLY]=[1]
#locator = MonthLocator(interval=2)
#fig, ax = plt.subplots()
#ax.plot(time,sim,'r', label='Sim')
#ax.plot(time,obs,'b', label='Obs')
legend = ax6.legend(loc='upper right', shadow=True)
##ax.set_xlim(0,270)

fig.suptitle('Daily Water Temperature of Deschutes (13A060)',fontsize=20)

#plt.locator_params(nbins=4)
start1, end1 = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start1, end1, 40))
start2, end2 = ax2.get_xlim()
ax2.xaxis.set_ticks(np.arange(start2, end2, 40))
start3, end3 = ax3.get_xlim()
ax3.xaxis.set_ticks(np.arange(start3, end3, 40))
start4, end4 = ax4.get_xlim()
ax4.xaxis.set_ticks(np.arange(start4, end4, 40))
start5, end5 = ax5.get_xlim()
ax5.xaxis.set_ticks(np.arange(start5, end5, 40))
start6, end6 = ax6.get_xlim()
ax6.xaxis.set_ticks(np.arange(start6, end6, 40))

pylab.ylim([11,22])
ax3.set_xlabel('time (day)',fontsize=18, x=1, y=0)
ax.set_ylabel('Water Temperature (degC)',fontsize=18)
matplotlib.rc('xtick', labelsize=14)
matplotlib.rc('ytick', labelsize=14)

bias=np.sum(abs(sim-obs))/data.shape[0]
tempr1=0
tempr2=0
tempr3=0
for i in range(0,data.shape[0]):
    tempr1+=(sim[i]-np.mean(sim))*(obs[i]-np.mean(obs))
    tempr2+=(sim[i]-np.mean(sim))**2
    tempr3+=(obs[i]-np.mean(obs))**2
    i+=1
R2=(tempr1/((tempr2*tempr3)**0.5))**2
RMSE=(sum((sim-obs)**2)/data.shape[0])**0.5
#print x
print bias,R2,RMSE
bias_1="{:.2f}".format(bias)
R2_1="{:.2f}".format(R2)
RMSE_1="{:.2f}".format(RMSE)
ax.text(730670,21,'$R^2= $'+str(R2_1),fontsize=14)
ax.text(730670,20.3,'$RMSE= $'+str(RMSE_1),fontsize=14)
ax.text(730670,19.6,'$Bias= $'+str(bias_1),fontsize=14)
#plt.show()
```
