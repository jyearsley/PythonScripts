```python
import matplotlib.pyplot as plt
from matplotlib.finance import quotes_historical_yahoo
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
import numpy as np
import pandas as pd
import datetime as dt
from matplotlib.dates import date2num, num2date 
from matplotlib.pyplot import figure, show
from matplotlib import scale as mscale
from matplotlib import transforms as mtransforms
import pylab
import matplotlib
import os
%matplotlib inline

os.chdir('/raid/caoqian/PugetSound/1_skagit/DHSVM3.1.2/output/12189500/11_current/00_06')
data1 = np.genfromtxt('flow.txt',dtype='int')
data1_1 = np.genfromtxt('flow.txt')
#data1 = np.genfromtxt('flow.txt',usecols=0)
year = data1[:,0]
month = data1[:,1]
day = data1[:,2]
sim = data1_1[:,3]
data2 = np.genfromtxt('obsflow.txt',dtype='int')
data2_1 = np.genfromtxt('obsflow.txt')
obs = data2_1[:,3]
data3 = np.genfromtxt('obsrain.txt',dtype='int')

data3_1 = np.genfromtxt('obsrain.txt')
sizetemp=data3_1.shape[0]-1
#print sizetemp
rain = data3_1[0:sizetemp,3]
time1=np.zeros([data1.shape[0],1])
for i in range(0,data1.shape[0]):
    a1= year[i]
    b1= month[i]
    c1= day[i]
    d1= dt.date(a1,b1,c1)
    time1[i] = date2num(d1)
    i+=1
#print time1

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.plot_date(time1, obs, 'b',label='Obs')
ax.plot_date(time1, sim, 'r',label='Sim')
pylab.ylim([0,1800])

ax2 = ax.twinx()
ax2.bar(time1,rain)
#ax2= plt.gca()
ax2.xaxis_date()
pylab.ylim([0,0.7])
ax2.invert_yaxis()
print np.max(time1), np.min(time1),np.size(rain),np.size(sim)

plt.gcf().autofmt_xdate()
legend = ax.legend(loc='center left', shadow=True)
plt.title('Daily flow in Skagit (12189500)',fontsize=16)

ax.set_xlabel('time (day)',fontsize=14, x=0.5, y=0)
ax.set_ylabel('Flow (cms)',fontsize=14)
ax2.set_ylabel('Rain(m/day)',fontsize=14)
matplotlib.rc('xtick', labelsize=14)
matplotlib.rc('ytick', labelsize=14)
ax.set_xlim([np.min(time1), np.max(time1)])

tempr1=0
tempr2=0
for i in range(0,sim.shape[0]):
    tempr1+=(sim[i]-obs[i])**2
    tempr2+=(obs[i]-np.mean(obs))**2
    i+=1
nash=1-tempr1/tempr2
nash_1="{:.2f}".format(nash)
ax.text(time1[20],1450,'$Nash= $'+str(nash_1),fontsize=16)
#ax2.set_xlim([np.max(time1), np.min(time1)])
#plt.show()
```
