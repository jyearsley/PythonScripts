**<font color='blue'>Read the input water quality files<br></font>
<span style="background-color: #FFFF00">NOTE THAT the input used here only consists of sim and obs records with matching timestamps. The script used to extract the matching records are in: C:\Users\Ning\Documents\IPython Notebooks\Write-Out-Matching-Records.ipynb</span>**

```python
basin_name = ['Mercer', 'Thornton', 'Issaquah']
basin_nameL = ['mercer', 'thornton', 'issaquah']
mainpath = {}
simsedfilepath = {}
simotherfilepath = {}
simTPfilepath = {}

# assign the file path
for i, n in enumerate(basin_name):
    mainpath[n] = 'D:\\Dropbox\\Python_Scripts\\'+str(basin_name[i])
    simsedfilepath[n] = str(mainpath[n])+ '\\sed_sim_obs.txt'
    simotherfilepath[n] = str(mainpath[n])+'\\coliform_sim_obs.txt'
    simTPfilepath[n] = str(mainpath[n])+'\\TP_sim_obs.txt'

# read data from input files
FsimdateJunk1 = collections.defaultdict(list)
FobsdateJunk1 = collections.defaultdict(list)
FsimdateJunk2 = collections.defaultdict(list)
FobsdateJunk2 = collections.defaultdict(list)
Fsimdate = collections.defaultdict(list)
Fobsdate = collections.defaultdict(list)

FsimSed = collections.defaultdict(list)
FobsSed = collections.defaultdict(list)
FsimOther = collections.defaultdict(list)

FobsOther = collections.defaultdict(list)
FsimTP = collections.defaultdict(list)
FobsTP = collections.defaultdict(list)

for i, n in enumerate(basin_name): 
    # date and time are stored in two separate columns, so here reading them together
    FsimdateJunk1[n] = np.genfromtxt(simsedfilepath[n], dtype=str, skiprows=0, usecols=[0,1])
    FobsdateJunk1[n] = np.genfromtxt(simsedfilepath[n], dtype=str, skiprows=0, usecols=[3,4])
    # print Fobsdate
    
    # read simulations
    FsimSed[n] = np.genfromtxt(simsedfilepath[n], dtype=float, skiprows=0, usecols=[2])
    FsimOther[n] = np.genfromtxt(simotherfilepath[n], dtype=float, skiprows=0, usecols=[2])
    FsimTP[n] = np.genfromtxt(simTPfilepath[n], dtype=float, skiprows=0, usecols=[2])
    
    # read observations
    FobsSed[n] = np.genfromtxt(simsedfilepath[n], dtype=float, skiprows=0, usecols=[5])
    FobsOther[n] = np.genfromtxt(simotherfilepath[n], dtype=float, skiprows=1, usecols=[5])
    FobsTP[n] = np.genfromtxt(simTPfilepath[n], dtype=float, skiprows=0, usecols=[5])

# stitch the date and time columns into one column
for i, n in enumerate(basin_name): 
    for j in range(len(FsimdateJunk1[n])):
        FsimdateJunk2[n].append(str(FsimdateJunk1[n][j][0])+'-'+str(FsimdateJunk1[n][j][1])) 
        FobsdateJunk2[n].append(str(FobsdateJunk1[n][j][0])+'-'+str(FobsdateJunk1[n][j][1]))

# convert the date/time string to python date/time format
for i, n in enumerate(basin_name):
    for j in range(len(FsimdateJunk2[n])):
        Fsimdate[n].append(datetime.datetime.strptime(FsimdateJunk2[n][j],'%Y-%m-%d-%H:%M:%S'))
        Fobsdate[n].append(datetime.datetime.strptime(FobsdateJunk2[n][j],'%Y-%m-%d-%H:%M:%S'))
        # debug
        if (i == 0 and j <=5):
            print Fsimdate[n][j], Fobsdate[n][j]
```

<font color='blue'>Plot the **Sim** vs. **obs** monthly **sediment conc**</font>

```python
ssim = {}
sobs = {}
junk = {}
FsimHighSed = collections.defaultdict(list)
FsimHighSedPre = collections.defaultdict(list)
FsimHighSedPost = collections.defaultdict(list)
FobsHighSed = collections.defaultdict(list)
calstartdate = datetime.datetime(1993,1,1)
calenddate = datetime.datetime(2002,1,1)
Fdate = collections.defaultdict(list)

print len(FsimSed['Mercer']), len(FobsSed['Mercer'])

for i, n in enumerate(basin_name):
    for j in range(len(FsimSed[n])):
        if (n == 'Mercer'): 
            threshold = 4
        if (n == 'Thornton'): 
            threshold = 6
        if (n == 'Issaquah'): 
            threshold = 4 
        if (FobsSed[n][j]>threshold and FobsSed[n][j]<1000):
            FsimHighSed[n].append(FsimSed[n][j])
            FsimHighSedPre[n].append(FsimSedPre[n][j])
            FsimHighSedPost[n].append(FsimSedPost[n][j])
            FobsHighSed[n].append(FobsSed[n][j])
            Fdate[n].append(Fsimdate[n][j])
            
for i, n in enumerate(basin_name):
    ssim[n] = pd.Series(FsimHighSed[n], index=Fdate[n])
    sobs[n] = pd.Series(FobsHighSed[n], index=Fdate[n])

fig=plt.figure(figsize=(20,15),dpi=400)

jj = 1
for i, n in enumerate(basin_name):
    ax = plt.subplot(3, 2, jj)
    jj = jj + 2
    # add spacing between subplots
    fig.subplots_adjust(hspace = 0.3, wspace = 0.1)
    
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    title(n, fontsize=14.0, y=1.02, fontweight="bold")
        
    if (n == 'Mercer'): 
        textstr1 = 'Calibration:\n$R^2 = 0.69$'
        plt.text(0.6, 0.9, textstr1, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        textstr2 = 'Validation:\n$R^2 = 0.54$'
        plt.text(0.4, 0.9, textstr2, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        threshold = 40
        plt.ylim([0, 150])

    if (n == 'Thornton'): 
        textstr1 = 'Calibration:\n$R^2 = 0.72$'
        plt.text(0.6, 0.9, textstr1, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        textstr2 = 'Validation:\n$R^2 = 0.56$'
        plt.text(0.4, 0.9, textstr2, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        threshold = 40
        plt.ylim([0, 350])
        
    if (n == 'Issaquah'): 
        textstr1 = 'Calibration:\n$R^2 = 0.64$'
        plt.text(0.6, 0.9, textstr1, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        textstr2 = 'Validation:\n$R^2 = 0.72$'
        plt.text(0.4, 0.9, textstr2, fontsize=14, transform=ax.transAxes, verticalalignment='top', bbox=props)
        threshold = 40
        plt.ylim([0, 700])

    ssim[n]['1993':'2009'].plot(label='Sim', color ='red', linewidth=2)
    sobs[n]['1993':'2009'].plot(label='Obs', color ='black', linewidth=1, style='-o', markersize=5)
    
    ax.axvspan(calstartdate, calenddate, alpha=0.5, color='0.85')
    
    # Plot the error bar for selected data (greater values than a threshold)
    c = []
    d = []
    e = []
    f = []
    for k in range(len(FsimSed[n])):
        if (FsimSed[n][k] > threshold or FobsSed[n][k] > threshold):
            f.append(Fsimdate[n][k])
            c.append(FsimSed[n][k])
            d.append(FsimSedPost[n][k]) 
            e.append(FsimSedPre[n][k])
    a = c - pd.DataFrame([d, c, e]).min()
    b = pd.DataFrame([d, c, e]).max() - c        
    #errorbar(f, c, yerr=[a, b], fmt=None, ecolor='0.55', lw=8, capsize=3, capthick=2)
    
    grid(True)
    if (i == 0):
        ax.legend(loc='upper left')
    if (i == 2):
        plt.xlabel('Year', fontsize=12., labelpad=20)
    plt.ylabel('conc. ($mg/l$)', fontsize=12.)

    # remove the tick label
    #ax.set_xticklabels([])

    # force all axis label placed horizontally
    plt.setp(ax.xaxis.get_majorticklabels(), rotation='horizontal')
    
    # set ticker properties
    ax.xaxis.set_tick_params(width=1, length=5)
    ax.get_xaxis().tick_bottom()
    
    
    
###########################################################################################################################################    
    
    ax = plt.subplot(3, 4, 3+4*i)
    # add spacing between subplots
    fig.subplots_adjust(hspace = 0.3, wspace = 0.15)
    nnn = len(sobs[n])
    d = np.linspace(1, nnn, num=nnn)/(nnn+1)
    y  = norm.ppf(d,0,1);
    
    # combined the data points within the 9-hour interval
    junk[n] = FsimHighSed[n] + FsimHighSedPre[n] + FsimHighSedPost[n]
    nn = len(junk[n])
    dd = np.linspace(1, nn, num=nn)/(nn+1)
    yy  = norm.ppf(dd,0,1);
    
    # create the axis ticks
    p  = [0.01, 0.05, 0.2, 0.5, 0.75, 0.90, 0.99];
    # relate var "p" with "y"
    tick  = norm.ppf(p,0,1);
    label = ['0.01','0.05','0.2','0.5','0.75','0.90','0.99'];
    
    # sort the data in an ascending order
    sobs[n].sort()
    junk[n].sort()
    
    # plot with simulation points that align with the obs
    ax.plot(yy, junk[n], 'r+', label='sim')
    ax.plot(y, sobs[n],'k*', label='obs')
    
    # use numpoints option so the markeres don't appear twice in one legend
    ax.xaxis.set_major_locator(ticker.FixedLocator(tick))
    ax.xaxis.set_major_formatter(ticker.FixedFormatter(label))
    
    if (i == 0):
        ax.legend(loc='best', numpoints = 1)
    if ( i == 2):
        plt.xlabel('Cumulative Probability', fontsize=12., labelpad=20)
        
    grid(True)

    
outputfilepath = 'D:\\Dropbox\\Python_Scripts\\sediment.png'    
savefig(str(outputfilepath),dpi=500,format = 'png',transparent='True')

```
