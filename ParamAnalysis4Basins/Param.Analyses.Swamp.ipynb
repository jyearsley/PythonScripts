```python

# Enter file path for sim. and obs. flow files
paramfile = 'D:\\Dropbox\\Python_Scripts\\ParamAnalysis\\allparam_swamp.txt'

# Read flow sim and obs dates &  data
ID = np.genfromtxt(paramfile, dtype=int, skiprows=1, usecols=[0])
H1 = np.genfromtxt(paramfile, skiprows=1, usecols=[4])
H2 = np.genfromtxt(paramfile, skiprows=1, usecols=[5])
POW1 = np.genfromtxt(paramfile, skiprows=1, usecols=[6])
POW2 = np.genfromtxt(paramfile, skiprows=1, usecols=[7])

sed_max = np.genfromtxt(paramfile, skiprows=1, usecols=[8])
sed_rate = np.genfromtxt(paramfile, skiprows=1, usecols=[9])
coliform_max = np.genfromtxt(paramfile, skiprows=1, usecols=[10])
coliform_rate= np.genfromtxt(paramfile, skiprows=1, usecols=[11])
coliform_W1 = np.genfromtxt(paramfile, skiprows=1, usecols=[12])
coliform_W2 = np.genfromtxt(paramfile, skiprows=1, usecols=[13])
r2_Sed = np.genfromtxt(paramfile, skiprows=1, usecols=[16])
r2_Col = np.genfromtxt(paramfile, skiprows=1, usecols=[18])
nash_sed = np.genfromtxt(paramfile, skiprows=1, usecols=[15])
nash_col= np.genfromtxt(paramfile, skiprows=1, usecols=[17])

# find the column that contains the max value
x = np.array(r2_Sed)
idx = x.argmax()
print "max_r2_sed = ", max(r2_Sed), "@ ID ", ID[idx]

x = np.array(nash_sed)
idx = x.argmax()
print "max_nash_sed = ", max(nash_sed), "@ ID ", ID[idx]

# find the column that contains the max value
x = np.array(r2_Col)
idx = x.argmax()
print "max_r2_coliform = ", max(r2_Col), "@ ID", ID[idx]

x = np.array(nash_col)
idx = x.argmax()
print "max_nash_coliform = ", max(nash_col), "@ ID", ID[idx]

print "Total parameter sets = ", len(H1)
```

Compute <b><span style="color:#f00">Correlation</span></b> between parameters and R2

```python
plt.figure(figsize=(7,7),dpi=400)
ax = plt.subplot(2,1,1)
plot(H1, r2_Sed, 'bo')
title('H1', fontsize=14., y=1.02, fontweight='bold')
ax = plt.subplot(2,1,2)
plt.hist(H1)
np.corrcoef(H1,r2_Sed)
```
