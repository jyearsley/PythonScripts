<font color='blue'>Read the sim and obs temperature files</font>

```python
gage_id = ['1170', '1070', '1060', '860', '840', '830', '780', '730', '710', '660', '650', '640', '570', '360']
rbm_id = ['279', '159', '280', '281', '166', '283', '287', '285', '159', '158', '289', '289', '291', '293']
scenarios = ['baseline_with_resv', 'full_with_resv', 'zero_with_resv', 'baseline_wo_resv', 'full_wo_resv', 'zero_wo_resv']
colors = ['#228b22', '#daa520', '#0000ff', 'black', 'r', '#7cfc00'] 
linestyles = ['-', '-', '-', '-', '-', '-']
strings = ['Base+Resvr' ,'Full+Resvr', 'Zero+Resvr', 'Base', 'Full', 'Zero']
mainpath = "D:\\Box Sync\\ConnecticutPaper1\\temperature_plots\\Ottau\\"

simtmppath = collections.defaultdict(dict)
obstmppath = collections.defaultdict(dict)
sim_tmp = collections.defaultdict(dict)
obs_tmp = collections.defaultdict(dict)
sim_tmp_date = collections.defaultdict(dict)
obs_tmp_date = collections.defaultdict(dict)

# Enter file path for sim. and obs. flow files
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        simtmppath[n][s] = str(mainpath)+str(s)+"\seg"+str(n)+".temp.txt"
    
for i, n in enumerate(gage_id):
    obstmppath[n] = str(mainpath)+"obs\\Ottau_"+str(n)+".txt"
```

```python
# Read flow sim and obs dates &  data
for i, n in enumerate(gage_id):
    obs_tmp_date[i] = np.genfromtxt(obstmppath[n], dtype=str, skiprows=0, usecols=[0])
    obs_tmp[i] = np.genfromtxt(obstmppath[n], dtype=float, skiprows=0, usecols=[1])
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        sim_tmp[i][j] = np.genfromtxt(simtmppath[n][s], dtype=float, skiprows=0, usecols=[1])
        sim_tmp_date[i][j] = np.genfromtxt(simtmppath[n][s], dtype=str, skiprows=0, usecols=[0])
```

<font color='blue'>Average the hourly or 3-hourly simulated temperature date into daily values.</font>

```python
simdate = collections.defaultdict(dict)
for i, n in enumerate(rbm_id):
    simdate[i] = collections.defaultdict(list)
obsdate = collections.defaultdict(list)
DailyPrecip = collections.defaultdict(dict)
DailySim = collections.defaultdict(dict)
DailyObs = collections.defaultdict(dict)
MonthlySim = collections.defaultdict(dict)
MonthlyObs = collections.defaultdict(dict)

for i, n in enumerate(gage_id):
    for j in range(len(obs_tmp_date[i])):
        obsdate[i].append(datetime.datetime.strptime(obs_tmp_date[i][j],'%m/%d/%Y-%H:%M:%S'))
        
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        for k in range(len(sim_tmp_date[i][j])):
            simdate[i][j].append(datetime.datetime.strptime(sim_tmp_date[i][j][k],'%m/%d/%Y-%H:%M'))
            
# daily mean flow    
for i,n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        DailySim[i][j] = pd.Series(sim_tmp[i][j], index=simdate[i][j]).resample('D', how='mean')
for i,n in enumerate(gage_id):
    DailyObs[i] = pd.Series(obs_tmp[i], index=obsdate[i]).resample('D', how='mean')
```

<font color='blue'> **Truncate** the obs. and sim. flow files based on dates.</font>

```python
startdate1 = datetime.datetime(2005,5,29) # for ottauquechee
enddate1 = datetime.datetime(2005,10,5)

startdate2 = datetime.datetime(2004,6,6) # for ottauquechee
enddate2 = datetime.datetime(2004,10,3)

# Truncate the data into desired time frame
for i, n in enumerate(gage_id):
    if (n=="1070" or n=="710"):
        DailyObs[i] = DailyObs[i].truncate(before=startdate2, after=enddate2)
        for j, s in enumerate(scenarios):
            DailySim[i][j] = DailySim[i][j].truncate(before=startdate2, after=enddate2)
    else:
        DailyObs[i] = DailyObs[i].truncate(before=startdate1, after=enddate1)
        for j, s in enumerate(scenarios):
            DailySim[i][j] = DailySim[i][j].truncate(before=startdate1, after=enddate1)
```

<font color='blue'> **Monthly** obs. and sim. flow.</font>

```python
# monthly mean flow for each year
for i,n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        MonthlySim[i][j] = DailySim[i][j].resample('M', how='mean')
for i,n in enumerate(rbm_id):
    MonthlyObs[i] = DailyObs[i].resample('M', how='mean')
    
# average monthly streamflow over the period
for i,n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        MonthlySim[i][j] = MonthlySim[i][j].groupby(lambda x: x.month).mean()
for i,n in enumerate(gage_id):
    MonthlyObs[i] = MonthlyObs[i].groupby(lambda x: x.month).mean()
```

<font color='blue'>Compute **Nash** between daily sim. and obs.</font>

```python
nash = collections.defaultdict(dict)
for i, n in enumerate(rbm_id):
    obs_mean= sum(DailyObs[i])/float(len(DailyObs[i]))
    ss_tot = sum((x-obs_mean)**2 for x in DailyObs[i]) 
    for j, s in enumerate(scenarios):
        sim_mean= sum(DailySim[i][j])/float(len(DailySim[i][j]))
        ss_err = sum((y-x)**2 for y,x in zip(DailySim[i][j], DailyObs[i]))
        nash[i][j] = 1 - (ss_err/ss_tot)
        # print "gage[%s] %s daily nash = %.2f" % (str(rbm_id[i]), str(s), nash[i][j])
```

<font color='blue'>Compute **R2** between daily sim. and obs. </font>

```python
R2 = collections.defaultdict(dict)
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        gradient, intercept, r_value, p_value, std_err = stats.linregress(DailySim[i][j],DailyObs[i])
        R2[i][j] = r_value ** 2
        # print "gage[%s] daily R2 = %.2f" % (str(rbm_id[i]), R2[i][j])
```

<font color='blue'>Compute **bias** between daily sim. and obs. </font>

```python
Bias = collections.defaultdict(dict)
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        tot_num = len(DailySim[i][j])
        Bias[i][j] = sum((y-x) for y,x in zip(DailySim[i][j], DailyObs[i]))/tot_num
        #print "gage[%s] daily Bias = %.2f" % (str(rbm_id[i]), Bias[i][j])
```

<font color='blue'>Compute **RMSE** between daily sim. and obs. </font>

```python
RMSE = collections.defaultdict(dict)
for i, n in enumerate(rbm_id):
    for j, s in enumerate(scenarios):
        tot_num = len(DailySim[i][j])
        RMSE[i][j] = (sum((y-x)**2 for y,x in zip(DailySim[i][j], DailyObs[i]))/tot_num) ** 0.5
        # print "gage[%s] daily RMSE = %.2f" % (str(rbm_id[i]), RMSE[i][j])
```
