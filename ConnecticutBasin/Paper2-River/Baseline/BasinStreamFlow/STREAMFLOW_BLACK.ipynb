```python
gage_id = ['01151500'] #', 01152500', '01153000', '01153550']
basin_name = ['Ottauquechee'] #, 'Sugar', 'Black', 'Williams']
mainpath = "D:\\Dropbox\\Python_Scripts\\\Conn_Streamflow\\black_ottau\\"
obsflowpath = {}
obsflowdata = {}
simflowdata = {}
tempsimflowdate = collections.defaultdict(list)
tempobsflowdate = collections.defaultdict(list)

# Enter file path for sim. and obs. flow files
simflowpath = str(mainpath) + 'Streamflow.Only'
for i, n in enumerate(gage_id):
    obsflowpath[n] = str(mainpath) + 'streamflow.' + str(n) + '.txt'
# Enter file path for observed precipitation files
precipfilepath = str(mainpath)+'Mass.Balance'

# Read flow sim and obs dates &  data
for i, n in enumerate(gage_id):
    simflowdata[n] = np.genfromtxt(simflowpath, dtype=float, skiprows=2, usecols=[i+1])

for i, n in enumerate(gage_id):
    obsflowdata[n] = np.genfromtxt(obsflowpath[n], dtype=float, usecols=[1])

tempsimflowdate = np.genfromtxt(simflowpath, dtype=str, skiprows=2, usecols=0)
for i, n in enumerate(gage_id):
    tempobsflowdate[n] = np.genfromtxt(obsflowpath[n], dtype=str, skiprows=0, usecols=0)

# Read obs. precipitation data
precipdata = np.genfromtxt(precipfilepath, dtype=float, skiprows=0, usecols=[10])
```

<font color='blue'>Average the 3-hourly simulated flow and precipitation into daily values.</font>

```python
simflowdate = []
obsflowdate = collections.defaultdict(list)
DailyPrecip = {}
DailySimFlow = {}
DailyObsFlow = {}

for i, n in enumerate(gage_id):
    for j in range(len(tempobsflowdate[n])):
        obsflowdate[n].append(datetime.datetime.strptime(tempobsflowdate[n][j],'%Y-%m-%d'))
    
for i in range(len(tempsimflowdate)):
    simflowdate.append(datetime.datetime.strptime(tempsimflowdate[i],'%m.%d.%Y-%H:%M:%S'))
    
for i,n in enumerate(gage_id):
    DailySimFlow[n] = pd.Series(simflowdata[n], index=simflowdate).resample('D', how='sum')/24/3600 #unit:m/s
    #DailyPrecip[n] = pd.Series(precipdata, index=simflowdate).resample('D', how='sum')  #unit: m/day
    DailyObsFlow[n] = pd.Series(obsflowdata[n], index=obsflowdate[n])
```

<font color='blue'>**Sort** and **match** the obs. and sim. flow files based on dates.</font>

```python
startdate1 = datetime.datetime(1979,10,1)
enddate1 = datetime.datetime(1989,9,30)
startdate2 = datetime.datetime(1996,10,1)
enddate2 = datetime.datetime(2001,9,30)

# Truncate the data into desired time frame
for i, n in enumerate(gage_id):
    if (n == '01153000'):
        DailySimFlow[n] = DailySimFlow[n].truncate(before=startdate1, after=enddate1)
        #DailyPrecip[n] = DailyPrecip[n].truncate(before=startdate1, after=enddate1)
        DailyObsFlow[n] = DailyObsFlow[n].truncate(before=startdate1, after=enddate1)
    else:
        DailySimFlow[n] = DailySimFlow[n].truncate(before=startdate2, after=enddate2)
        #DailyPrecip[n] = DailyPrecip[n].truncate(before=startdate2, after=enddate2)
        DailyObsFlow[n] = DailyObsFlow[n].truncate(before=startdate2, after=enddate2)
```
