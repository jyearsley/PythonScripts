```python
import matplotlib.pyplot as plt
from matplotlib.finance import quotes_historical_yahoo
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
import numpy as np
import pandas as pd
import datetime as dt
from matplotlib.dates import date2num, num2date 
from matplotlib.pyplot import figure, show
from matplotlib.backends.backend_pdf import PdfPages

filepath='1_1.txt'
#dates = dt.datetime.strptime("07/27/2012","%m/%d/%Y")
#print dates

data = np.loadtxt(filepath)
data1 = np.genfromtxt(filepath,dtype='int')
print data1
year = data1[:,0]
month = data1[:,1]
day = data1[:,2]
sim = data[:,3]
obs = data[:,4]

#The shape attribute for numpy arrays returns the dimensions of 
#the array. If Y has n rows and m columns, then Y.shape is (n,m). 
#So Y.shape[0] is n.
time=np.zeros([data.shape[0],1])
```

```python
#print len(time)
#print len(sim)
for i in range(0,data.shape[0]):
    a= year[i]
    b= month[i]
    c= day[i]
    d = dt.date(a,b,c)
    time[i] = date2num(d)
    i+=1
time1=np.arange(data.shape[0])
fig, ax = plt.subplots()
ax.plot(time1,sim,'r', label='Sim')
ax.plot(time1,obs,'b', label='Obs')
legend = ax.legend(loc='upper right', shadow=True)
#ax.set_xlim(0,270)
plt.xlabel('time',fontsize=18)
plt.ylabel('Water Temperature (degC)',fontsize=18)
plt.title('Daily Water Temperature of Deschutes',fontsize=20, y=1.05)
plt.show()
```
