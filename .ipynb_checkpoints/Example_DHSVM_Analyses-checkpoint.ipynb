###This notebook includes Python tools for analyzing DHSVM outputs.
Time series data are analyzed with **numpy** and **pandas** tools and plotted with **matplotlib**. Spatial maps are analyzed and plotted using **netCDF4** and **Basemap** utilities. In general, the scripts that needs to be changed (eg. file path, locations, time range, and variables to plot, etc) are on top of the coding block if there's any. However, differences in the input data may require changing the code, such as which columns to read as the dates, how many rows to skip as the header, etc. Currently all input file paths are absolute paths so you should be able to run the code as it is and plot from my output files. All plots are saved into a pdf page (specify the file name in the second code block). Note that you'll have to run the last code block to close the pdf file you opened in order to view the file correctly. 

```python
# Import all needed libraries 
import sys, os, datetime, time
import numpy as np
import pandas as pd
import matplotlib.dates as dates
import matplotlib.ticker as ticker
from matplotlib.backends.backend_pdf import PdfPages
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.mplot3d import Axes3D
from netCDF4 import Dataset
```

First, various **hydrographs** are created from DHSVM output file **"Streamflow.Only"**. The first code block reads the header of the file and the time range, then return the column names and time range. 

```python
# Initialize a pdf file for saving all the figures
pdfpages = PdfPages('30m/temp_0.005/hydrograph.pdf')

# Enter file path for simulation results
infile = '/nfs/therapy/raid/tcchiao/Bull_DHSVM/output/30m/temp_0.005/Streamflow.Only'
# Enter file path for observed stream flow files
obspath = '/nfs/therapy/raid/tcchiao/Bull_obsflow/1976_2011/'
rivers = ['LITTLE_SANDY','NORTH_FORK', 'FIR_CREEK', 'SOUTH_FORK', 'BULL_RUN']




# Read the first line in the infile and return each word
all_loc = open(infile,'r').readline().split()
close(infile)
# Read the first column in the infile as the dates
tempdates = np.genfromtxt(infile, dtype = str, skiprows = 2, usecols = 0)
# Convert the string format dates into a datetime object
simdates = []
for i in range(len(tempdates)):
    simdates.append(datetime.datetime.strptime(tempdates[i],'%m.%d.%Y-%H:%M:%S'))
print 'Stream flow data from locations:',all_loc[1:], 'are available for plotting for time ranging from', simdates[0], 'to', simdates[-1]
```

This code block asks for the column names for which data should be loaded, then proceed to open the text file and load the data. The loaded data has almost the same shape as it is in the text file. That is, each column in the text file (ie. each stream flow saver location) would be in a column in the DataFrame, with each time step being the row index. 

```python
# Input the locations you'd like to process data for
locations = ['BULL_RUN_ADD', 'BULL_RUN', 'NORTH_FORK', 'FIR_CREEK_ADD', 'FIR_CREEK', 'SOUTH_FORK', 'SOUTH_FORK_ADD', 'OUTLET', 'LITTLE_SANDY', 'UPPER_LITTLE_SANDY']




# Map each location to the column number of the corresponding data
locs = map(all_loc.index,locations)
# Load the simulation results from the corresponding column into a pandas dataframe object and use the dates as the index
# 
sim = {}
for loc, location in zip(locs,locations):
    sim[location] = np.loadtxt(infile, skiprows = 2, usecols = [loc])
# Average the flow by day and convert from m^3 per time step (3hr) to per second
sim = pd.DataFrame(sim, index=simdates).resample('D')/3/3600

# Load the observation data into a pandas dataframe object and use the dates as the index
obs = {}
for i,river in enumerate(rivers):
    obs[river] = np.loadtxt(obspath+river, skiprows =27, usecols = [3])

# Load the dates of the observation data, input the correct column number 
tempdates = np.loadtxt(obspath+river, dtype = str, skiprows =27, usecols = [2])

# Convert the string format dates into a datetime object
obsdates = []
for i in range(len(tempdates)):
    obsdates.append(datetime.datetime.strptime(tempdates[i],'%Y-%m-%d'))
    
# Average the flow by day and convert from m^3 per time step (3hr) to per second
obs = pd.DataFrame(obs, index=obsdates).resample('D')*0.0283168
```

This code block is for any additional processing of the stream flow data to re-create the flow accumulation areas of USGS gages. In my case, some columns have to be combined for a direct comparison with the gage data. Comment out if unnecessary 

```python
# Combine flows to compare with observed data
# Note that the operation replace the original flow with the combined flow, 
# Thus this block should not be run repeatedly without repeating the code block above! 
sim['BULL_RUN'] = sim['BULL_RUN'] + sim['BULL_RUN_ADD']
sim['FIR_CREEK'] = sim['FIR_CREEK'] + sim['FIR_CREEK_ADD']
sim['SOUTH_FORK'] = sim['SOUTH_FORK'] + sim['SOUTH_FORK_ADD']
sim['LITTLE_SANDY'] = sim['LITTLE_SANDY'] + sim['UPPER_LITTLE_SANDY']

# Delete the unused flows
for key in ['BULL_RUN_ADD','FIR_CREEK_ADD','SOUTH_FORK_ADD','UPPER_LITTLE_SANDY','OUTLET']:
    del sim[key]
# Sum all flows to total
sim['TOTAL'] = sim.sum(axis=1)
obs['TOTAL'] = obs.sum(axis=1)
```

The four code blocks below plots four different kind of hydrographs that might be of interest. Each block asks for the location and time range input to be plotted. Each hydrograph is saved into the pdf file you specify above. 

```python
# Input locations to plot, comment out if the same as data to be processed
locations = ['LITTLE_SANDY','NORTH_FORK', 'FIR_CREEK', 'SOUTH_FORK', 'BULL_RUN','TOTAL']
# Enter time range you'd like to plot, date format should be year, month, date 
startdate = datetime.datetime(1982,10,1)
enddate = datetime.datetime(1985,9,30)




# Truncate the data into desired time frame
sim2plot = sim.truncate(before=startdate, after=enddate)
obs2plot = obs.truncate(before=startdate, after=enddate)

# Create time series plot of the daily averaged flow at each location
figure(figsize=(15,20))
suptitle('Daily Averaged Stream Flow', fontsize = 20., y=1.02)
for i,location in enumerate(locations):
    subplot(len(locations),1,i)
    # Plot simulated flow of the desired date range
    sim2plot[location].plot(label = 'Simulated Flow', color = 'b')
    # Plot observed flow of the desired date range
    obs2plot[location].plot(label = 'Observed Flow', color = 'r')
    # Titles, axes labels and ticks, units, grid
    title('Daily Averaged Stream Flow at ' + location, fontsize=12.0)
    xlabel('Time', fontsize = 12.0)
    ylabel(r'$Stream flow (m^3/s)$', fontsize = 15.0)
    grid(True)
    tight_layout()
    legend(loc='best')
    
# Save the output as separate figures
#savefig('dailyflow.png',dpi=300,format = 'png',transparent='True')
#savefig('dailyflow.pdf',dpi=300,format = 'pdf')
# Save the figure into the multipage pdf file
pdfpages.savefig()
```

```python
# Input locations to plot, comment out if the same as data to be processed
locations = ['LITTLE_SANDY','NORTH_FORK', 'FIR_CREEK', 'SOUTH_FORK', 'BULL_RUN','TOTAL']
# Enter time range you'd like to plot, date format should be year, month, date 
startdate = datetime.datetime(1982,10,1)
enddate = datetime.datetime(1985,9,30)




# Create time series plot of the cumulative flow at each location
# Calculate cumulative sum 
obscum = obs2plot.cumsum()
simcum = sim2plot.cumsum()

# Plot
figure(figsize=(15,15))
suptitle('Cumulative Flow', fontsize = 20., y=1.02)
for i,location in enumerate(locations):
    subplot(len(locations),1,i)
    # Plot simulated flow of the desired date range
    simcum[location].plot(label = 'Simulated Flow', color = 'b')
    # Plot observed flow of the desired date range
    obscum[location].plot(label = 'Observed Flow', color = 'r')
    # Titles, axes labels and ticks, units, grid
    title('Cumulative flow at ' + location, fontsize=12.0)
    ylabel(r'$Stream flow (m^3/s)$', fontsize = 15.0)
    grid(True)
    tight_layout()
    legend(loc='best')
    
# Save the output as separate figures
#savefig('dailyflow.png',dpi=300,format = 'png',transparent='True')
#savefig('dailyflow.pdf',dpi=300,format = 'pdf')
# Save the figure into the multipage pdf file
pdfpages.savefig()
```

```python
# Create annual cumulative flow plots at each location according to the wateryear
# Input locations to plot, comment out if the same as data to be processed
locations = ['LITTLE_SANDY','NORTH_FORK', 'FIR_CREEK', 'SOUTH_FORK', 'BULL_RUN','TOTAL']
# Enter time range you'd like to plot, date format should be year, month, date 
startdate = datetime.datetime(1982,10,1)
enddate = datetime.datetime(1985,9,30)




# A simple function to define the water year of a date
def wateryear(inyear):
    if inyear.month >= 10:
        return inyear.year+1
    else:
        return inyear.year

# Group data in obs and sim based by its wateryear
obsyear = obs2plot.groupby(lambda x: wateryear(x))
simyear = sim2plot.groupby(lambda x: wateryear(x))

# For each wateryear, calculate the cumulative sum and then append to a big data package
for i, (year,group) in enumerate(obsyear):
    if i == 0:
        obsyearcum = group.cumsum()
    else:
        obsyearcum = obsyearcum.append(group.cumsum())
for i, (year,group) in enumerate(simyear):
    if i == 0:
        simyearcum = group.cumsum()
    else:
        simyearcum = simyearcum.append(group.cumsum())
    
# Plot
figure(figsize=(15,15))
suptitle('Annual Cumulative Flow', fontsize = 20., y=1.02)
for i,location in enumerate(locations):
    subplot(len(locations),1,i)
    # Plot simulated flow of the desired date range
    simyearcum[location].plot(label = 'Simulated Flow', color = 'b')
    # Plot observed flow of the desired date range
    obsyearcum[location].plot(label = 'Observed Flow', color = 'r')
   # Titles, axes labels and ticks, units, grid
    title('Annual Cumulative Flow at ' + location, fontsize=12.0)
    ylabel(r'$Stream flow (m^3/s)$', fontsize = 15.0)
    grid(True)
    tight_layout()
    legend(loc='best')
    
# Save the output as separate figures
#savefig('dailyflow.png',dpi=300,format = 'png',transparent='True')
#savefig('dailyflow.pdf',dpi=300,format = 'pdf')
# Save the figure into the multipage pdf file
pdfpages.savefig()
```

Before moving onto other DHSVM output files, calculate various **index values** to judge the fit of simulated flow to observed flow on a daily basis. 

```python
# Calculate Nash Coefficient
obsave = obs2plot.mean()
diff1 = obs2plot - sim2plot
diff2 = obs2plot - obsave
nash = 1 - ((diff1*diff1).sum()/(diff2*diff2).sum())
print 'Nash Coefficient at each location is'
print nash
print

# Calculate Mean Absolute Percentage Error
mape = (diff1/obs2plot).abs().mean()*100
print 'Mean Absolute Percentage Error (MAPE) at each location is'
print mape
print 

# Calculate Root Mean Square Error
mse = (diff1*diff1).mean()
print 'Root Mean Square Error (RMSE) at each location is'
for i,location in enumerate(locations):
    rmse = sqrt(mse[location])
    print location, '   ', rmse
print 'TOTAL', '    ', sqrt(mse['TOTAL'])
print 

# Calculate Absolute Bias
bias = diff1.abs().mean()
print 'Absolute Bias at each location is'
print bias
```
