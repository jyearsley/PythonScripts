<font color='blue'>Read the input **streamflow** files (change the path if needed)</font>

```python
test_id = range(1, 16)
print test_id[8:]
```

```python
basin_name = ['Ottauquechee'] 
mainpath = "D:\\Box Sync\\Connecticut_Basin\\Ottau-Calibration-Results\\"
obsflowpath = {}
obsflowdata = {}
simflowdata = {}
simflowpath = {}
tempsimflowdate = collections.defaultdict(list)
tempobsflowdate = collections.defaultdict(list)

# Enter file path for sim. and obs. flow files
for i, n in enumerate(test_id):
    simflowpath[n] = str(mainpath)+'test'+str(n)+'\\Streamflow.Only'
    
obsflowpath = str(mainpath)+'streamflow.01151500.txt'

# Read flow sim and obs dates &  data
for i, n in enumerate(test_id):
    simflowdata[n] = np.genfromtxt(simflowpath[n], dtype=float, skiprows=2, usecols=[1])
    tempsimflowdate[n] = np.genfromtxt(simflowpath[n], dtype=str, skiprows=2, usecols=[0])
    
obsflowdata = np.genfromtxt(obsflowpath, dtype=float, usecols=[1])
tempobsflowdate = np.genfromtxt(obsflowpath, dtype=str, skiprows=0, usecols=[0])
```

<font color='blue'>Average the 3-hourly simulated flow into daily values.</font>

```python
simflowdate = collections.defaultdict(list)
obsflowdate = []
DailyPrecip = {}
DailySimFlow = {}
DailyObsFlow = {}
MonthlySimFlow = {}
MonthlyObsFlow = {}

for j in range(len(tempobsflowdate)):
    obsflowdate.append(datetime.datetime.strptime(tempobsflowdate[j],'%Y-%m-%d'))

for i, n in enumerate(test_id):
    for j in range(len(tempsimflowdate[n])):
        simflowdate[n].append(datetime.datetime.strptime(tempsimflowdate[n][j],'%m.%d.%Y-%H:%M:%S'))

# daily mean flow    
for i,n in enumerate(test_id):
    DailySimFlow[n] = pd.Series(simflowdata[n], index=simflowdate[n]).resample('D', how='sum')/24/3600 #unit:m/s

DailyObsFlow = pd.Series(obsflowdata, index=obsflowdate)
```

<font color='blue'> **Truncate** the obs. and sim. flow files based on dates.</font>

```python
startdate1 = datetime.datetime(1950,10,1) # for ottauquechee
# enddate1 = datetime.datetime(1957,9,29)
enddate1 = datetime.datetime(1953,9,29)

# Truncate the data into desired time frame
for i, n in enumerate(test_id):
    DailySimFlow[n] = DailySimFlow[n].truncate(before=startdate1, after=enddate1)
DailyObsFlow = DailyObsFlow.truncate(before=startdate1, after=enddate1)
```

<font color='blue'> **Monthly** obs. and sim. flow.</font>

```python
# monthly mean flow for each year
for i,n in enumerate(test_id):
    MonthlySimFlow[n] = DailySimFlow[n].resample('M', how='mean')
MonthlyObsFlow = DailyObsFlow.resample('M', how='mean')
    
# average monthly streamflow over the period
for i,n in enumerate(test_id):
    MonthlySimFlow[n] = MonthlySimFlow[n].groupby(lambda x: x.month).mean()
MonthlyObsFlow = MonthlyObsFlow.groupby(lambda x: x.month).mean()
```

<font color='blue'>Compute **Nash** between sim. and obs. **daily** flow </font>

```python
nash_flow = {}
obs_mean= sum(DailyObsFlow)/float(len(DailyObsFlow))
ss_tot = sum((x-obs_mean)**2 for x in DailyObsFlow) 
for i, n in enumerate(test_id):
    sim_mean= sum(DailySimFlow[n])/float(len(DailySimFlow[n]))
    ss_err = sum((y-x)**2 for y,x in zip(DailySimFlow[n], DailyObsFlow))
    nash_flow[n] = 1 - (ss_err/ss_tot)
    print "test[%s] nash = %.2f" % (str(test_id[i]), nash_flow[n])
```

<font color='blue'>Plot sim. and obs. flow</font>

```python
ss = 3.5 * len(test_id)
fig=plt.figure(figsize=(13,ss),dpi=400)
jj = 1
for i, n in enumerate(test_id[8:]):
    ax = plt.subplot2grid((len(test_id), 3), (0+i, 0), colspan=2)
    fig.subplots_adjust(hspace = 0.6, wspace = 0.4)
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(.02,.9, 'Test'+str(n), fontsize=12, horizontalalignment='left', transform=ax.transAxes)
    
    textstr1 = 'Nash = %.2f' % nash_flow[n]
    ax.text(.02,.7, textstr1, fontsize=12, color='red',horizontalalignment='left', transform=ax.transAxes)
    
    DailyObsFlow.plot(label='Obs', color='grey', linewidth=2, style='-', alpha=0.8)
    DailySimFlow[n].plot(label='Sim', color='blue', linewidth=2, style='-', alpha=0.8)
    ax.grid(True)
    if (i == 0):
        leg=ax.legend(loc=1, numpoints = 1)
        leg.get_frame().set_alpha(0.5)
    plt.ylabel('flow ($m^{3}/s$)', fontsize=12., labelpad=5)
    
    #################### Cumulative Distribution Function ####################
    jj = jj + 1
    ax = plt.subplot2grid((len(test_id), 3), (0+i, 2), colspan=1)
    jj = jj + 1
    
    nn = len(DailySimFlow[n])
    # log-normal probability plot
    d = np.linspace(1, nn, num=nn)/(nn+1)
    y = norm.ppf(d, 0, 1)
        
    # create the axis ticks
    p  = [0.01, 0.2, 0.5, 0.90, 0.999];
    
    # relate var "p" with "y"
    tick  = norm.ppf(p,0,1);
    label = ['0.01','0.2','0.5','0.90','0.999'];
    
    # sort the data in an ascending order
    DailySimFlow[n].sort()
    DailyObsFlow.sort()
    ax.plot(y, DailyObsFlow,'*', color = '0.9', alpha = 0.8, label='obs')
    ax.plot(y, DailySimFlow[n], 'b+', label='sim')
    
    # use numpoints option so the markeres don't appear twice in one legend
    ax.xaxis.set_major_locator(ticker.FixedLocator(tick))
    ax.xaxis.set_major_formatter(ticker.FixedFormatter(label))
    ax.grid(True)
    
    # Changing the label's font-size
    ax.tick_params(axis='x', labelsize=11)
    
    if (i == 0):
        leg=ax.legend(loc='best', numpoints = 1)
        leg.get_frame().set_alpha(0.5)
    plt.ylabel('flow ($m^{3}/s$)', fontsize=12., labelpad=5)
    if (i == 1):
        plt.xlabel('probability', fontsize=12., labelpad=5)
    
outfig = 'D:\\Box Sync\\Connecticut_Basin\\Ottau-Calibration-Results\\JunkFlow.png'    
#plt.savefig(str(outfig),dpi=500,format = 'png',transparent='True')
```
