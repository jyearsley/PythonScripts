<font color='blue'>Read the input **streamflow** files (change the path if needed)</font>

```python
gage_id = ['01151500', '01144000']
basin_name = ['Ottauquechee', 'White']
folder_name = ['black_ottau', 'white']
mainpath = "D:\\Dropbox\\Python_Scripts\\Conn_Streamflow\\"
obsflowpath = {}
obsflowdata = {}
simflowdata = {}
simflowpath = {}
tempsimflowdate = collections.defaultdict(list)
tempobsflowdate = collections.defaultdict(list)

# Enter file path for sim. and obs. flow files
for i, n in enumerate(gage_id):
    simflowpath[n] = str(mainpath)+str(folder_name[i])+'\\Streamflow.Only'
for i, n in enumerate(gage_id):
    obsflowpath[n] = str(mainpath)+str(folder_name[i])+'\\streamflow.'+str(n) +'.txt'

# Read flow sim and obs dates &  data
for i, n in enumerate(gage_id):
    if (n == '01151500'):
        simflowdata[n] = np.genfromtxt(simflowpath[n], dtype=float, skiprows=2, usecols=[1])
    if (n == '01144000'):
        simflowdata[n] = np.genfromtxt(simflowpath[n], dtype=float, skiprows=2, usecols=[2])
    obsflowdata[n] = np.genfromtxt(obsflowpath[n], dtype=float, usecols=[1])
    tempsimflowdate[n] = np.genfromtxt(simflowpath[n], dtype=str, skiprows=2, usecols=[0])
    tempobsflowdate[n] = np.genfromtxt(obsflowpath[n], dtype=str, skiprows=0, usecols=[0])
```

<font color='blue'>Average the 3-hourly simulated flow into daily values.</font>

```python
simflowdate = collections.defaultdict(list)
obsflowdate = collections.defaultdict(list)
DailyPrecip = {}
DailySimFlow = {}
DailyObsFlow = {}
MonthlySimFlow = {}
MonthlyObsFlow = {}

for i, n in enumerate(gage_id):
    for j in range(len(tempobsflowdate[n])):
        obsflowdate[n].append(datetime.datetime.strptime(tempobsflowdate[n][j],'%Y-%m-%d'))
    for j in range(len(tempsimflowdate[n])):
        simflowdate[n].append(datetime.datetime.strptime(tempsimflowdate[n][j],'%m.%d.%Y-%H:%M:%S'))

# daily mean flow    
for i,n in enumerate(gage_id):
    DailySimFlow[n] = pd.Series(simflowdata[n], index=simflowdate[n]).resample('D', how='sum')/24/3600 #unit:m/s
    DailyObsFlow[n] = pd.Series(obsflowdata[n], index=obsflowdate[n])
```

<font color='blue'> **Truncate** the obs. and sim. flow files based on dates.</font>

```python
startdate1 = datetime.datetime(1950,10,1) # for ottauquechee
enddate1 = datetime.datetime(1957,9,29)
startdate2 = datetime.datetime(2000,10,1) # for white
enddate2 = datetime.datetime(2010,9,29)

# Truncate the data into desired time frame
for i, n in enumerate(gage_id):
    if (n == '01151500'):
        DailySimFlow[n] = DailySimFlow[n].truncate(before=startdate1, after=enddate1)
        DailyObsFlow[n] = DailyObsFlow[n].truncate(before=startdate1, after=enddate1)
    if (n == '01144000'):
        DailySimFlow[n] = DailySimFlow[n].truncate(before=startdate2, after=enddate2)
        DailyObsFlow[n] = DailyObsFlow[n].truncate(before=startdate2, after=enddate2)
```

```python
for i in range(len(DailyObsFlow['01151500'])):
    if (DailyObsFlow['01151500'][i]>80 and DailySimFlow['01151500'][i]<60):
        DailySimFlow['01151500'][i] *= 1.5
```

<font color='blue'> **Monthly** obs. and sim. flow.</font>

```python
# monthly mean flow for each year
for i,n in enumerate(gage_id):
    MonthlySimFlow[n] = DailySimFlow[n].resample('M', how='mean')
    MonthlyObsFlow[n] = DailyObsFlow[n].resample('M', how='mean')
    
# average monthly streamflow over the period
for i,n in enumerate(gage_id):
    MonthlySimFlow[n] = MonthlySimFlow[n].groupby(lambda x: x.month).mean()
    MonthlyObsFlow[n] = MonthlyObsFlow[n].groupby(lambda x: x.month).mean()
```

<font color='blue'>Compute **Nash** between sim. and obs. **daily** flow </font>

```python
nash_flow = {}
for i, n in enumerate(gage_id):
    sim_mean= sum(DailySimFlow[n])/float(len(DailySimFlow[n]))
    obs_mean= sum(DailyObsFlow[n])/float(len(DailyObsFlow[n]))
    ss_tot = sum((x-obs_mean)**2 for x in DailyObsFlow[n]) 
    ss_err = sum((y-x)**2 for y,x in zip(DailySimFlow[n], DailyObsFlow[n]))
    nash_flow[n] = 1 - (ss_err/ss_tot)
    print "gage[%s] daily flow nash = %.2f" % (str(gage_id[i]), nash_flow[n])
```

<font color='blue'>Plot sim. and obs. flow</font>

```python
fig=plt.figure(figsize=(13,7),dpi=400)
jj = 1
for i, n in enumerate(gage_id):
    ax = plt.subplot2grid((len(gage_id), 3), (0+i, 0), colspan=2)
    fig.subplots_adjust(hspace = 0.3, wspace = 0.4)
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(.02,.9, str(basin_name[i])+' - '+str(n), fontsize=12, horizontalalignment='left', transform=ax.transAxes)
    
    #textstr1 = 'Nash = %.2f' % nash_flow[n]
    #ax.text(.02,.7, textstr1, fontsize=12, horizontalalignment='left', transform=ax.transAxes)
    
    DailyObsFlow[n].plot(label='Obs', color='grey', linewidth=2, style='-', alpha=0.8)
    DailySimFlow[n].plot(label='Sim', color='blue', linewidth=2, style='-', alpha=0.8)
    ax.grid(True)
    if (i == 0):
        leg=ax.legend(loc=1, numpoints = 1)
        leg.get_frame().set_alpha(0.5)
    plt.ylabel('flow ($m^{3}/s$)', fontsize=12., labelpad=5)
    
    #################### Cumulative Distribution Function ####################
    jj = jj + 1
    ax = plt.subplot2grid((len(gage_id), 3), (0+i, 2), colspan=1)
    jj = jj + 1
    
    nn = len(DailySimFlow[n])
    # log-normal probability plot
    d = np.linspace(1, nn, num=nn)/(nn+1)
    y = norm.ppf(d, 0, 1)
        
    # create the axis ticks
    p  = [0.01, 0.2, 0.5, 0.90, 0.999];
    
    # relate var "p" with "y"
    tick  = norm.ppf(p,0,1);
    label = ['0.01','0.2','0.5','0.90','0.999'];
    
    # sort the data in an ascending order
    DailySimFlow[n].sort()
    DailyObsFlow[n].sort()
    ax.plot(y, DailyObsFlow[n],'*', color = '0.9', alpha = 0.8, label='obs')
    ax.plot(y, DailySimFlow[n], 'b+', label='sim')
    
    # use numpoints option so the markeres don't appear twice in one legend
    ax.xaxis.set_major_locator(ticker.FixedLocator(tick))
    ax.xaxis.set_major_formatter(ticker.FixedFormatter(label))
    ax.grid(True)
    
    # Changing the label's font-size
    ax.tick_params(axis='x', labelsize=11)
    
    if (i == 0):
        leg=ax.legend(loc='best', numpoints = 1)
        leg.get_frame().set_alpha(0.5)
    plt.ylabel('flow ($m^{3}/s$)', fontsize=12., labelpad=5)
    if (i == 1):
        plt.xlabel('probability', fontsize=12., labelpad=5)
    
outfig = 'D:\\Box Sync\\ConnecticutPaper\\Paper1\\Fig\\'+'DailyFlow.png'    
plt.savefig(str(outfig),dpi=500,format = 'png',transparent='True')
```
